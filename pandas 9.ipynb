# pivot table

```python
# pivot allows you to trnsform or reset the data
```

```python
import pandas as pd
df = pd.read_csv('weather9.csv')
df
```

```python
df.pivot(index='dates',columns ='city')
```

```python
df.pivot(index='dates',columns ='city',values = 'humidity')
```

```python
df.pivot(index='humidity',columns ='city')
```

pivot tble is used to summarize and aggregate data inside dataframe

```python
df = pd.read_csv('weather9_part2.csv')
df
```

```python
df.pivot_table(index = 'city',columns = 'dates') # it will give average of humidity andd temp
```

```python
df.pivot_table(index = 'city',columns = 'dates',aggfunc = 'sum')
```

```python
df.pivot_table(index = 'city',columns = 'dates',aggfunc = 'count')
```

```python
df.pivot_table(index = 'city',columns = 'dates',aggfunc = 'diff')
```

```python
df.pivot_table(index = 'city',columns = 'dates',aggfunc = 'mean') # default
```

```python
df.pivot_table(index = 'city',columns = 'dates',margins = 'True')
```

# grouper in pivot table

```python
df = pd.read_csv('weather9_part3.csv')
```

```python
df
```

```python
# we wil use freq = month as we want to find avg of month
# first we have to convert in datetime index
# first convert string object into date
```

```python
df['dates'] = pd.to_datetime(df['dates'])
```

```python
type(df['dates'])
```

```python
df.pivot_table(index = pd.Grouper(freq='M',key='dates'),columns = 'city')
```
