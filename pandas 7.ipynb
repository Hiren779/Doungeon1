# concatt datafrm

```python
import pandas as pd
india_weather = pd.DataFrame({
    'city':['mumbai','delhi','banglore'],
    'temperature' : [32,45,30],
    'humidity' : [80,60,78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    'city':['new york','chicago','orlando'],
    'temperature' : [21,14,35],
    'humidity' : [68,65,75]
})
us_weather
```

```python
df = pd.concat([india_weather,us_weather])
df
```

```python
df = pd.concat([india_weather,us_weather],ignore_index = True)
df
```

```python
df = pd.concat([india_weather,us_weather],keys = ['india','us']) # doesnt work with ignore index
df
# created additional index
```

```python
df.loc['india']
```

```python
df.loc['us']
```

```python
temperature_df = pd.DataFrame({
       'city':['mumbai','delhi','banglore'],
        'temperature' : [32,45,30]    
})
temperature_df
```

```python
windspeed_df = pd.DataFrame({
        'city':['mumbai','delhi','banglore'],
        'windspeed' : [9,12,7]    
})
windspeed_df
```

```python
df = pd.concat([temperature_df,windspeed_df])
df 
# it create additional row which is not accepted
```

```python
# by default axis = 0 so it append data set as rows
```

```python
df = pd.concat([temperature_df,windspeed_df],axis = 1)
df 
```

```python
# lets say order of city is differet and city is missing
```

```python
windspeed_df = pd.DataFrame({
        'city':['delhi','mumbai'],
        'windspeed' : [9,12]    
})
windspeed_df
```

```python
# so lets use index
temperature_df = pd.DataFrame({
       'city':['mumbai','delhi','banglore'],
        'temperature' : [32,45,30],
   
}, index = [0,1,2])
temperature_df
```

```python
windspeed_df = pd.DataFrame({
        'city':['delhi','mumbai'],
        'windspeed' : [9,12]    
},index = [1,0])
windspeed_df
```

```python
df = pd.concat([temperature_df,windspeed_df],axis = 1) #now it is runnng okay
df 
```
