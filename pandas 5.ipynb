# handle missing data replace function

```python
import pandas as pd  #-999 ,-88is empty value
import numpy as np
df = pd.read_csv('weather-data5.csv')
df
```

```python
new_df = df.replace(-99999,np.NaN)
new_df
```

```python
new_df = df.replace([-99999,-88888],np.NaN) # replace these two value if has -88888
new_df
```

```python
new_df = df.replace({'temperature':-99999,
                    'windspeed':-99999,
                    'event':0
                    },np.NaN) # replce value column vise -99999 in first both and zero in event
new_df
```

```python
# convert no event t zero in file
```

```python
new_df = df.replace({
    -99999:np.NaN,
    'no event':'Sunny'
})
new_df
```

```python
# we want to chop off miles per hour or unit
```

```python
new_df = df.replace('[A-Za-z]','',regex=True) # as it also replace event column
new_df
```

```python
# when windspeed is 6 mph, 7 mph etc. & temperature is 32 F, 28 F etc.
new_df = df.replace({'temperature': '[A-Za-z]', 'windspeed': '[a-z]'},'', regex=True) 
new_df
```

# Replacing list with another list
