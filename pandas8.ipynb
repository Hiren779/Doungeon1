# merge dataframe

```python
import pandas as pd
df1 = pd.DataFrame({
    'city':['new york','chicago','orlando'],
    'temperature':[21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    'city':['chicago','new york','orlando'],
    'humidity':[65,68,75],
})
df2
```

```python
df3 = pd.merge(df1,df2,on='city') # not matching by index it is matching by city
df3
```

```python
df1 = pd.DataFrame({
    'city':['new york','chicago','orlando','baltina'],
    'temperature':[21,14,35,32],
})
df1
```

```python
df2 = pd.DataFrame({
    'city':['chicago','new york','san fransisco'],
    'humidity':[65,68,71],
})
df2
```

```python
df3 = pd.merge(df1,df2,on='city') # only merge common cities
df3 # it uses intersection
```

```python
df3 = pd.merge(df1,df2,on='city',how = 'outer') #  how,outer argument represent union
df3 
```

```python
df3 = pd.merge(df1,df2,on='city',how = 'inner') #  inner joint is default
df3 
```

```python
df3 = pd.merge(df1,df2,on='city',how = 'left') # takeall value f left dataframe as here df1 is left and common element
df3
```

```python
df3 = pd.merge(df1,df2,on='city',how = 'right')
df3
```

```python
df3 = pd.merge(df1,df2,on='city',how = 'outer',indicator=True) # says from where it has ome
df3
```

```python
df1 = pd.DataFrame({
    'city':['new york','chicago','orlando','baltina'],
    'temperature':[21,14,35,32],
    'humidity' : [65,68,75,71]
})
df1
```

```python
df2 = pd.DataFrame({
    'city':['chicago','new york','san diego'],
    'temperature':[21,14,35],
    'humidity' : [65,68,71]
})
df2
```

```python
df3 = pd.merge(df1,df2,on='city') # _x means it has appended suffixes
df3
```

```python
df3 = pd.merge(df1,df2,on='city',suffixes = ('_left','_right'))
df3
```
